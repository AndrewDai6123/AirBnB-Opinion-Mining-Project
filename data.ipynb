```python
# Given the column we will have a function that does some text preprocessing such as making all lowercase, removing stop words and any punctuation
def pre_processing(text):
    
    # Get the stopwords we desire in english
    stoplist = set(stopwords.words("english"))
    
    punctuation_list = string.punctuation
    
    # remove all puncutations in the reviews
    removed_punctuations = [char for char in text if char not in punctuation_list]
    
    removed_punctuations = ''.join(removed_punctuations)
    
    #print(removed_punctuations)

    remove_stopwords = [words for words in removed_punctuations.split() if words.lower() not in stoplist]
    
    #print(remove_stopwords)
    
    return ' '.join(remove_stopwords)
```

```python
# Initalize a new list to hold the preprocessed comments values to add to a new column Rewviews
# takes about 2 mins to run 

def clean_reviews(dataframe):
    
    commentList = []

    # For every row in reviews2_df iterate it through to get the values
    for index, row in dataframe.iterrows():
        
        #print(index)
        
        # Get the comment from the comment column
        comment = row['comments']
        
        #print("Original Comment")
        #print(comment)    
        
        # Apply the Pre-processing Function to the comment value
        process_comment = pre_processing(comment)
        
        # Append the processed comment to the list
        commentList.append(process_comment)
        
        #print("Pre-process Comment")
        #print(process_comment)
        
    #print(commentList)
    # Initalize a new column
    dataframe["Reviews"] = commentList
```

```python
for i in range(len(reviews_list)):
    clean_reviews(reviews_list[i])
```
